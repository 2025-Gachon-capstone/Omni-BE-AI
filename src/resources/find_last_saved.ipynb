```python
import pandas as pd

df = pd.read_csv('master_dataset.csv')
print(df.head(5))  # 또는 df.head()만 해도 기본 5줄 출력
```

```python
# user_id가 숫자형(int)인 경우
max_user_id = df['user_id'].max()
print(f"user_id의 최대값은 {max_user_id}입니다.")
```

```python

user_id = 49804
matching_rows = df[df['user_id'] == user_id]

if not matching_rows.empty:
    row_index = matching_rows.index[0]  # 첫 번째로 일치하는 인덱스
    print(f"user_id {user_id}는 DataFrame의 {row_index + 2}번째 CSV 행에 있습니다.")  
else:
    print("해당 user_id를 찾을 수 없습니다.")
```

```python
csv_line_number = 289033
row_index = csv_line_number - 2  # 헤더 + 0-based 인덱스 보정

if 0 <= row_index < len(df):
    print(df.iloc[row_index])
else:
    print("해당 행 번호는 범위를 벗어났습니다.")
```
